```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def varphi_Y_N(N, t):
    return (np.sqrt(N)*(np.exp(t / (2*np.sqrt(N))) - np.exp(-t / (2*np.sqrt(N)))) / t)**N
```

```python
def varphi_G(t, sigma_square):
    return np.exp(sigma_square * t**2 / 2)
```

```python
# Plot \varphi_{Y_N}(t) and \varphi_{G}(t)
sigma_square = 1. / 12.
t = np.linspace(0.001, 5, num=1000)
plt.figure(figsize=(7,6))
for N in [1, 2, 5, 10]:
    y1 = varphi_Y_N(N, t)
    plt.plot(t, y1, label=r"$\varphi_{Y_{%d}}$" % N)

y2 = varphi_G(t, sigma_square)
plt.plot(t, y2, label=r"$\varphi_{G}$")
plt.legend(prop={'size': 20})
plt.xlabel("t")
plt.ylabel("y")
plt.savefig('p2b.png')
plt.show()
```

```python
# (c) Chernoff bound
def markov(u, N):
    return (N**2 * (np.exp(2*u/N**2) - np.exp(-2*u/N**2)) / (4*u*np.exp(4*u**2/N**2)))**N
```

```python
# Chebsyshev inequality
def chebsyshev(u, N):
    return 1./(12. * N * u**2)
```

```python
# Plot upper bounds as a function of u
u = np.linspace(0.001, 2, num=1000)
N = 1
y1 = markov(u, N)
y2 = chebsyshev(u, N)
plt.figure(figsize=(8,6))
plt.plot(u, y1, label="Markov")
plt.plot(u, y2, label="Chebsyshev")
plt.legend(prop={'size': 20})
plt.xlabel("u")
plt.ylabel("Upper Bound")
plt.ylim([0,1.2])
plt.title(f"N={N}")
plt.savefig('p2c_N_1.png')
plt.show()
```
