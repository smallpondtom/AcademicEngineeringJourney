```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
from scipy.stats import multivariate_normal
```

```python
mu1 = np.array([-1, 1])
mu2 = np.array([1, 0])
```

```python
sigma1 = np.array([[3, -6],
                   [-6, 24]])
sigma2 = np.array([[16, -6],
                   [-6, 8]])
```

```python
def F(x):
    return multivariate_normal.pdf(x, mean=mu1, cov=sigma1)/(multivariate_normal.pdf(x, mean=mu1, cov=sigma1) + multivariate_normal.pdf(x, mean=mu2, cov=sigma2))
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-20, 20, 100)

X, Y = np.meshgrid(x, y)
```

```python
Z = np.zeros((100,100))
```

```python
for i in range(100):
    for j in range(100):
        Z[i,j] = F([X[i,j], Y[i,j]])
```

```python
plt.contour(X, Y, Z, [0.5], colors='k')
plt.text(-7.5, 0, r"$\Gamma_2$", size=30)
plt.text(5.0, 0, r"$\Gamma_2$", size=30)
plt.text(-1.5, 12, r"$\Gamma_1$", size=30)
plt.text(-1.5, -12, r"$\Gamma_1$", size=30)
plt.savefig('p2.png')
plt.show()
```
