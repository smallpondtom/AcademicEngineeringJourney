```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return x**2 + y**2 + x*y + x + 3*y
```

```python
x = np.linspace(1./3. - 1, 1./3. + 1, 100)
y = np.linspace(-5./3. - 1, -5./3. + 1, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
#(c) contour plot of f(x)
cp = plt.contour(X, Y, Z, 30)
plt.colorbar(cp)
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('p2c.png')
plt.show()
```

```python
#(d) gradient descent
H = np.array([[2, 1],[1, 2]])
b = np.array([-1, -3])
x = np.zeros(2)
trace_x = np.zeros((2, 5))
for t in range(1, 5):
    r = b - np.dot(H, x)
    alpha = np.dot(r, r) / np.dot(r, np.dot(H, r))
    x += alpha * r
    trace_x[:, t] = x
```

```python
x = np.linspace(1./3. - 2, 1./3. + 2, 100)
y = np.linspace(-5./3. - 2, -5./3. + 2, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
cp = plt.contour(X, Y, Z, 30)
plt.colorbar(cp)
plt.plot(trace_x[0,:], trace_x[1, :], marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('p2d.png')
plt.show()
```

```python
#(e) conjugate gradient
H = np.array([[2, 1],[1, 2]])
b = np.array([-1, -3])
x = np.zeros(2)
r = b - np.dot(H, x)
d = b - np.dot(H, x)
trace_x = np.zeros((2, 3))
for t in range(1, 3):
    rTr = np.dot(r, r)
    alpha = rTr / np.dot(d, np.dot(H, d))
    x += alpha * d
    trace_x[:, t] = x
    r -= alpha * np.dot(H, d)
    beta = np.dot(r, r) / rTr
    d = r + beta * d
```

```python
x = np.linspace(1./3. - 2, 1./3. + 2, 100)
y = np.linspace(-5./3. - 2, -5./3. + 2, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
cp = plt.contour(X, Y, Z, 30)
plt.colorbar(cp)
plt.plot(trace_x[0,:], trace_x[1, :], marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('p2e.png')
plt.show()
```
