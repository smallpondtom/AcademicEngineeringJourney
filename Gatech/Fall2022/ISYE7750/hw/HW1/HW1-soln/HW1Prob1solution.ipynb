# Problem 1: Simulating quadratic functions

The arguments of the function are called $x_1$ and $x_2$ here instead of $w_1$ and $w_2$.

```python
from numpy import arange, array
from pylab import meshgrid
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt

# we'll plot quadratic functions of two variables x1 and x2
def z_func(x1,x2, A): #A here is expected to be a two-by-two multidimensional array
    
    value = x1 * x1 * A[0][0] + x2 * x2 * A[1][1] + x2 * x1 * A[1][0] + x1 * x2 * A[0][1]
    return value
    
x1 = arange(-3.0,3.0,0.1)
x2 = arange(-3.0,3.0,0.1)
X1,X2 = meshgrid(x1, x2) # grid of points

A = array([[1, 0],[0, 1]]) #Change this to simulate different loss functions
Z = z_func(X1, X2, A) # evaluation of the function on the grid
```

```python
#Code for 3D plotting

fig1 = plt.figure()
ax = fig1.gca(projection='3d')
surf = ax.plot_surface(X1, X2, Z, rstride=1, cstride=1, 
                      cmap=cm.RdBu,linewidth=0, antialiased=False)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig1.colorbar(surf, shrink=0.5, aspect=5)
```

```python
#Code for contour plotting

fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.set_aspect(1)
ax2.contourf(x1, x2, Z)
```

# Comment on the contour plot
<font color='blue'>
    First, the contours are circular and this is because the eigenvalues of $\mathbf{A}$ are same. Second, the region near the original point is dark and the region away from the original point is light. Correspondingly, in the 3D plot, the function value decreases when it goes close to the center and increases when we move away.</font>
